# Um terremoto de magnitude 7,8 atingiu o Equador em 16 de abril de 2016. As pessoas se reuniram nos esforÃ§os de socorro doando Ã¡gua e outros produtos de primeira necessidade, o que afetou muito as vendas dos supermercados por vÃ¡rias semanas apÃ³s o terremoto.


# OrientaÃ§Ã£o ChatGPT
Alternativa: Colocar uma escala de impacto (caso o efeito seja gradual):
Se vocÃª achar que o impacto nas vendas nÃ£o Ã© abrupto e sim gradual, pode considerar usar valores diferentes para o impacto. Ao invÃ©s de apenas 0 ou 1, vocÃª poderia criar uma escala de impacto, como:

0: Antes do terremoto.
1: Durante o terremoto e imediato apÃ³s o evento.
0.5: Em um perÃ­odo intermediÃ¡rio de recuperaÃ§Ã£o.
0: ApÃ³s o impacto significativo ter passado.
Isso poderia representar uma tendÃªncia de diminuiÃ§Ã£o do impacto ao longo do tempo.

```python
# Definindo o inÃ­cio e fim do impacto (16 de abril de 2016 atÃ© 16 de julho de 2016)
start_date = pd.to_datetime('2016-04-16')
end_date = start_date + pd.DateOffset(months=3)  # TrÃªs meses apÃ³s o terremoto

# FunÃ§Ã£o para verificar se a data estÃ¡ no perÃ­odo de impacto
def is_earthquake_impact(date):
    if start_date <= date <= end_date:
        return 1
    else:
        return 0
    


# Criando a nova coluna 'earthquake_impact'
loja_salles10['earthquake_impact'] = loja_salles10['date'].apply(is_earthquake_impact)
```
