```python
# # Orientação ChatGPT
# Função para verificar se o dia é o 15 ou o último dia do mês
def is_payment_day(date):
    if date.day == 15 or date == date.replace(day=1) + pd.DateOffset(months=1, days=-1):
        return 1
    else:
        return 0

# Aplicando a função para criar a nova coluna
loja_salles10['payment_day'] = loja_salles10['date'].apply(is_payment_day)
```

# Orientação ChatGPT
Alternativa: Colocar uma escala de impacto (caso o efeito seja gradual):
Se você achar que o impacto nas vendas não é abrupto e sim gradual, pode considerar usar valores diferentes para o impacto. Ao invés de apenas 0 ou 1, você poderia criar uma escala de impacto, como:

0: Antes do terremoto.
1: Durante o terremoto e imediato após o evento.
0.5: Em um período intermediário de recuperação.
0: Após o impacto significativo ter passado.
Isso poderia representar uma tendência de diminuição do impacto ao longo do tempo.

```python
# Definindo o início e fim do impacto (16 de abril de 2016 até 16 de julho de 2016)
start_date = pd.to_datetime('2016-04-16')
end_date = start_date + pd.DateOffset(months=3)  # Três meses após o terremoto

# Função para verificar se a data está no período de impacto
def is_earthquake_impact(date):
    if start_date <= date <= end_date:
        return 1
    else:
        return 0
    


# Criando a nova coluna 'earthquake_impact'
loja_salles10['earthquake_impact'] = loja_salles10['date'].apply(is_earthquake_impact)
```

```python
def aplicando_hiper_parametros(modelo, parametros, X_train, y_train, X_teste, y_teste):

  

        # Criando o GridSearchCV
        grid_search = RandomizedSearchCV(modelo, parametros, cv=3, scoring='r2', n_jobs=1)

        # Treinando o GridSearchCV
        grid_search.fit(X_train, y_train)

        # Pegando o melhor modelo encontrado e seu melhor parametro, para possivel analise e comparação
        melhor_modelo = grid_search.best_estimator_
        melhores_parametros = grid_search.best_params_

        # Fazendo previsões 
        y_pred = melhor_modelo.predict(X_teste)

        # Calculando R² de treino e teste
        acuracia_train = round(melhor_modelo.score(X_train, y_train) * 100, 2)
        acuracia_test = round(melhor_modelo.score(X_teste, y_teste) * 100, 2)

        print("Melhores parâmetros:", melhores_parametros)
        print("Acuracia de treino:", acuracia_train)
        print("Acuracia de teste:", acuracia_test)



    #return melhor_modelo, y_pred, melhores_parametros
```
